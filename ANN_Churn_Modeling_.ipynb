```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("D:\Churn_Modelling.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
data.describe()
```

```python
one_hot_encoding = pd.get_dummies(data, columns = ["Geography","Gender"])
```

```python
one_hot_encoding.head()
```

```python
one_hot_encoding.drop("Geography_France",axis = 1, inplace = True)
one_hot_encoding.drop("Gender_Female",axis = 1, inplace = True)
```

```python
one_hot_encoding.dtypes
```

```python
for colum in ["Geography_Germany","Geography_Spain","Gender_Male"]:
    one_hot_encoding[colum].replace({True : 1,False : 0},inplace=True)
```

```python
to=["RowNumber","CustomerId"]
one_hot_encoding.drop(to, axis = 1, inplace= True)
```

```python
one_hot_encoding
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = one_hot_encoding.drop("Exited",axis=1,inplace = False)
```

```python
one_hot_encoding
```

```python
x.drop("Surname",axis=1,inplace=True)
```

```python
y = one_hot_encoding["Exited"]
```

```python
y
```

```python
x_train, x_test, y_train ,y_test = train_test_split(x, y, test_size = 0.3, random_state = 75)
```

```python
from sklearn.preprocessing import StandardScaler
'''Normalization based on the Z-score value'''
```

```python
x_train = StandardScaler().fit_transform(x_train)
```

```python
x_test = StandardScaler().fit_transform(x_test)
```

```python
x_test
```

```python
x_train
```

```python
x_train.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LeakyReLU, ReLU, Dropout
'''
Sequential -- Able to Forward and Backward Propagation
Dense -- Creating Nodes, Input_Layer, Hidden_Layers and the output
Activation Funtion -- 
Dropout -- For handling Overfitting
'''
```

```python
model = Sequential()
model.add(Dense(units = 11))
'''
First Layer
'''
```

```python
model.add(Dense(units = 12,activation = 'ReLU'))
'''
Hidden_Layer_1
'''
```

```python
model.add(Dense(units = 7,activation = 'LeakyReLU'))
'''model.add(Dropout(.3))'''
'''
Hidden_Layer_2
'''
```

```python
Sequential().add(Dense(units = 1,activation = 'sigmoid'))
'''model.add(Dropout(.3))'''
'''
Output_Layer
'''
```

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
'''
Optimezer
Loss_Function
Metrics
'''
```

```python
model_out = model.fit(x_train,y_train,epochs=40,batch_size=50, validation_split = 0.3)
```

```python
model_out.history.keys()
```

```python
plt.plot(model_out.history["accuracy"],color= "blue", linewidth = 1, label = "Traning Acu")
plt.plot(model_out.history["val_accuracy"], color = "orange", label = "Validation Acu")
plt.legend(loc="upper left" )  # Adjust location as needed
plt.xlabel("epochs")
plt.ylabel('accuracy')
```

```python
plt.plot(model_out.history["loss"],color= "red", label = "Traning loss")
plt.plot(model_out.history["val_loss"], color = "green", label = "Validation loss")
plt.legend()
plt.xlabel("epochs")
plt.ylabel('loss')
```

```python
import tensorflow
eraly_stoping = tensorflow.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0.001,
    patience=3,
    verbose=0,
    mode="min",
    baseline=None,
    restore_best_weights=False,
)
'''
Early Stopping
'''
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, ReLU, LeakyReLU
new_model = Sequential()
new_model.add(Dense(units = 11, activation = "ReLU"))
new_model.add(Dense(units = 11, activation = "LeakyReLU"))
new_model.add(Dense(units = 7, activation = "ReLU"))
new_model.add(Dense(units = 1, activation = "sigmoid"))
new_model.compile(optimizer = 'adam',loss = 'binary_crossentropy', metrics = ["mse"])
```

```python
new_out = new_model.fit(x_train,y_train, validation_split = .3, epochs = 1000, callbacks=eraly_stoping, batch_size = 50)
```

```python
new_out.history.keys()
```

```python
plt.plot(new_out.history["loss"], label = "Loss")
plt.plot(new_out.history["val_loss"], label = "Val_Loss")
plt.legend()
plt.xlabel("epochs")
plt.ylabel("loss")
```

```python
plt.plot(new_out.history["mse"], label = "MSE")
plt.plot(new_out.history["val_mse"], label = "Val_MSE")
plt.legend()
plt.xlabel("epochs")
plt.ylabel("accuracy mse")
```

```python
y_pred = new_model.predict(x_test)
y_pred = y_pred >= 0.5
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
data = confusion_matrix(y_test,y_pred)
data
```

```python
accuracy = accuracy_score(y_test,y_pred)
accuracy
```
